# Построение нелинейным МНК логистической функции

Шаг 1. 
Задаем функцию fun1 для логистической модели, предсказывающих количество особей $y$ по номеру суток $t$.

````python
def fun1(t, a, b, c):
    return a/(1+b*np.exp(-c*t))
````

Методом подбора находим начальные коэффициенты:

````python
coef1 = [1600, 70, -0.8]
````

Используя функцию curve_fit, находим нужные коэффициенты для самой функции и строим график на основе данных таблицы для Rhiz:

````python
params1, _ = optimize.curve_fit(fun1,x,y1,coef1)
a1, b1, c1 = params1
print("Found: ", a1, b1, c1)
````

````python
arrayarg = np.linspace(np.min(x), np.max(x), 30)
plt.plot(x,y1,'o', color = 'blue')
plt.plot(arrayarg, fun1(arrayarg, a1, b1, c1), '-', color='red', label = 'fun1')
plt.title('Аппроксимация данных Rhiz:', size = 18)
plt.xlabel('Дней')
plt.ylabel('Кол-во особоей')
plt.legend()
plt.grid()
pass
````

Шаг 2:
Так же методом подбора выбираем начальные коэффициенты для функции со значениями от таблицы Sit:

````python
coef2 = [600, 500, -0.1]
````

Получаем окончательные коэффициенты при помощи функции curve_fit, затем строим график:

````python
params2, _ = optimize.curve_fit(fun1,x,y2,coef2)
a2, b2, c2 = params2
print("Found: ", a2, b2, c2)
````

````python
arrayarg = np.linspace(np.min(x), np.max(x), 50)
plt.plot(x,y2,'o', color = 'blue')
plt.plot(arrayarg, fun1(arrayarg, a2, b2, c2), '-', color='red', label = 'fun2')
plt.title('Аппроксимация данных Sit:', size = 20)
plt.xlabel('Дней')
plt.ylabel('Кол-во особоей')
plt.legend()
plt.grid()
pass
````

# Построение и сравнение решения из статьи с найденным решением.

Шаг 3: 
Теперь задаем функцию Rhiz для решения, данного нам из статьи:

````python
def fun3(t):
    return 338/(1+87*np.exp(-0.05*t))
````

Строим график аппроксимации этой функции красным цветом и подставляем ранее полученый на шаге 1 график аппроксимации Rhiz синим цветом:

````python
arrayarg = np.linspace(np.min(x), np.max(x), 30)
plt.plot(x,y1,'o', color = 'blue')
plt.plot(arrayarg, fun1(arrayarg, a1, b1, c1), '-', color='red', label = 'fun1')
plt.plot(arrayarg, fun3(arrayarg), '-', color='blue', label = 'fun2')
plt.title('Аппроксимация 2х ф-й НМНК для Rhiz', size = 14)
plt.xlabel('Дней')
plt.ylabel('Кол-во особоей')
plt.legend()
plt.grid()
pass
````

Как видно из полученного графика 2х аппроксимаций, аппроксимация Rhiz полученная в 1 шаге очень близко схожа по значениям с готовым решенением из статьи, но имеет совсем небольшие отклонения.

Шаг 4: 
    Задаем функцию Sit для готового решения из статьи:

````python
def fun4(t):
    return 200/(1+257*np.exp(-0.1*t))
````

Строим график аппроксимации полученной функций Sit желтым цветом и подставляем график аппроксимации Sit, полученный на шаге 2, красным цветом:

````python
arrayarg = np.linspace(np.min(x), np.max(x), 30)
plt.plot(x, y2,'o', color = 'blue')
plt.plot(arrayarg,fun1(arrayarg, a2, b2, c2), '-', color = 'red', label = 'fun2')
plt.plot(arrayarg, fun4(arrayarg), '-', color = 'yellow', label = 'fun4')
plt.title('Аппроксимация 2х ф-й НМНК для Sit', size = 14)
plt.xlabel('Дней')
plt.ylabel('Кол-во особей')
plt.legend()
plt.grid()
pass
````

Из данного графика видно, что график аппроксимации полученный на шаге 2 стал заметно отличаться по значению от готового решения, приведенного в статье, но тем не менее, так же близко отражает реальную картину аппроксимации.
